```python
import pandas as pd
import numpy as np
from progressbar import ProgressBar
```

```python
df = pd.read_csv('amazon_alexa.tsv',sep='\t')
```

```python
df.tail()
```

```python
df['rating'].value_counts()
```

```python

#in case there are blank reviews
df.dropna(inplace=True)

blanks = []  # start with an empty list

for i,rt,dt,vr,rv,fb in df.itertuples():
    if type(rv)==str:            
        if rv.isspace():
            blanks.append(i)     

df.drop(blanks, inplace=True)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
df['scores'] = df['verified_reviews'].apply(lambda review: sid.polarity_scores(review))

df['compound']  = df['scores'].apply(lambda score_dict: score_dict['compound'])

df['predicted'] = df['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
```

```python
df['actual'] = np.zeros(len(df['rating']))
```

```python
pbar = ProgressBar()
for i in pbar(range(0,3150)):
    if i in blanks:
        continue
    elif df['rating'][i] <= 3:
        df['actual'][i] = 'neg'
    else:
        df['actual'][i] = 'pos'
```

```python
df[df['predicted']=='neg']
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(confusion_matrix(df['actual'],df['predicted']))
```

```python
print(classification_report(df['actual'],df['predicted']))
```

```python
print(accuracy_score(df['actual'],df['predicted']))
```
